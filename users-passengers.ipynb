```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mysql.connector   
from pandas import DataFrame
```

```python
# объединим таблицы USERS и PASSENGERS

cnx = mysql.connector.connect(user='root', password='',
                              host='192.168.100.59',
                              database='preferences')

try:
   cursor = cnx.cursor()
   cursor.execute("""
      SELECT
        l.ID as ID,
        l.USER_INFO as USER_INFO,
        l.CREATEDAT as CREATED_USER,
        r.ID  as PASSENGER_ID,
        r.PASSENGER as PASSENGER,
        r.CREATED_AT as CREATED_PASSENGER

        From 
            USERS as l
        LEFT JOIN
            PASSENGERS as r
        ON l.ID = r.USER_ID
  
   """)
   result = cursor.fetchall()
 
finally:
    cnx.close()
```

```python
users_passengers = DataFrame(result)
```

```python
users_passengers
```

```python
users_passengers = users_passengers.rename(columns={ \
    0: 'USER_ID', 
    1: 'USER_INFO', 
    2: 'CREATED_USER', 
    3: 'PASSENGER_ID', 
    4: 'PASSENGER_INFO',
    5:'CREATED_PASSENGER'                        
    })
```

```python
count_passengers = users_passengers.groupby('USER_ID', as_index=False).agg({'PASSENGER_ID':'count'})
```

```python
# топ-15 юзеров по количеству пассажиров
count_passengers_top15 = count_passengers.sort_values(by = 'PASSENGER_ID', ascending=False).head(15)
count_passengers_top15
```
